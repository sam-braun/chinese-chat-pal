```python
import requests as req
import time

api_key = 'AjqDOaTFAm1qxTherm3JbH3djF3cIUlg' #insert API key
topic = 'China'
articles = []
year = 2006
month = 8

def contains_term(article, term):
    for key, value in article.items():
        if value and term.lower() in value.lower():
            return True
    return False
    

while True:
    try:
        
        while (year == 2006):
            while (month <= 12):
                url = 'https://api.nytimes.com/svc/archive/v1/' + str(year) + '/' + str(month) + '.json?api-key=' + api_key
                response = req.get(url).json()
                # print(response)

                # Extract necessary fields from the response
                docs = response['response']['docs']
                for doc in docs:
                    filteredDoc = {}
                    filteredDoc['headline'] = doc['headline']['main']
                    filteredDoc['pub_date'] = doc['pub_date']
                    # filteredDoc['keywords'] = doc['keywords']['name']
                    filteredDoc['news_desk'] = doc['news_desk']
                    filteredDoc['section_name'] = doc['section_name']
                    filteredDoc['abstract'] = doc['abstract']
                    filteredDoc['paragraph'] = doc['lead_paragraph']

                    if contains_term(filteredDoc, topic):
                        articles.append(filteredDoc)

                # time.sleep(6) # Avoid hitting API request limit
                month += 1
            month = 1
            year += 1

    except:
        print(response)
        break
# articles[:10]
```

```python
import csv

input_file = art_csv
output_file = "/Users/samuelbraun/Desktop/article2.csv"

with open(input_file, 'r') as csv_file:
    reader = csv.DictReader(csv_file)
    fieldnames = reader.fieldnames
    
    fieldnames.extend(['pub_cal_date', 'pub_time', 'original_pub_date_format'])

    rows = []
    for row in reader:
        # Split address on comma

        pub_cal_date, pub_time = row['pub_date'].rsplit('T', 1)
        row['pub_cal_date'] = pub_cal_date.strip()
        row['pub_time'] = pub_time.strip()
        row['pub_time'] = row['pub_time'][:-5]
        row['original_pub_date_format'] = row['pub_date']
        del row['pub_date']
        rows.append(row)

    fieldnames.remove('pub_date')
    
    new_order = ["headline", "pub_cal_date", "pub_time", "news_desk", "section_name", "abstract", "paragraph", "original_pub_date_format"]

# Write the processed rows into the new CSV file
with open(output_file, 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=new_order)
    writer.writeheader()
    for row in rows:
        writer.writerow(row)
```
