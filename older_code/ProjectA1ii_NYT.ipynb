```python
df = pd.read_csv("/Users/samuelbraun/Desktop/SummerPython/ProjectA1_NYT/article6.csv")
df = df[['headline', 'abstract', 'paragraph', 'pub_cal_date']]
df['full_article'] = df['headline'] + ' ' + df['abstract'] + ' ' + df['paragraph'] + ' ' + df['pub_cal_date']
df['full_article'] = df['full_article'].str.lower()
df.to_csv(("/Users/samuelbraun/Desktop/SummerPython/ProjectA1_NYT/article6.csv"))
df
```

```python
import string

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['full_article'] = df['full_article'].fillna('')
df['cln_full_article'] = df['full_article'].apply(remove_punctuation)

df
```

```python
from nltk.tokenize import word_tokenize

df['tk_cln_full_article'] = df['cln_full_article'].apply(word_tokenize)

df
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stop_words(tokens):
    return [word for word in tokens if word not in stop_words]

df['flt_tk_cln_full_article'] = df['tk_cln_full_article'].apply(remove_stop_words)

df
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    word_list = word_tokenize(text)
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w, pos='v') for w in word_list]) # 'v' stands for verb
    return lemmatized_output


# Apply the function to your DataFrame
# Ensure that 'flt_tk_cln_full_article' contains lists of words, not strings.
# If it's a list of words:
df['flt_tk_cln_full_article'] = df['flt_tk_cln_full_article'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# If it's a string already, there's no need to join.


df['lm_flt_tk_cln_full_article'] = df['flt_tk_cln_full_article'].apply(lemmatize_text)

df
```

```python
# Print the first 10 rows of the original and lemmatized texts
print(df[['flt_tk_cln_full_article', 'lm_flt_tk_cln_full_article']].head(10))

sample_text = 'The dogs were running quickly through the fields.'
print(lemmatize_text(sample_text))

```
