```python
import pandas as pd
df = pd.read_csv("/Users/samuelbraun/Desktop/SummerPython/ProjectA1_NYT/article7.csv")
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df['lm_flt_tk_cln_full_article'] = df['lm_flt_tk_cln_full_article'].fillna('')

vectorizer = TfidfVectorizer(max_features=5000)  # limiting to 5000 features for performance
tfidf_matrix = vectorizer.fit_transform(df['lm_flt_tk_cln_full_article'])

# The 'tfidf_matrix' now contains a numerical representation of your text data.
```

```python
ner_df = pd.DataFrame()

import spacy

spacy.prefer_gpu()
nlp = spacy.load('en_core_web_sm', disable=['tagger', 'parser'])


def extract_entities(docs):
    return [(ent.text, ent.label_) for doc in docs for ent in doc.ents]

docs = list(nlp.pipe(df['lm_flt_tk_cln_full_article'].values))
ner_df['entities'] = extract_entities(docs)


# def extract_entities(text):
#     doc = nlp(text)
#     return [(ent.text, ent.label_) for ent in doc.ents]

# # Apply the function to your DataFrame
# df['entities'] = df['lm_flt_tk_cln_full_article'].apply(extract_entities)

# df
```
